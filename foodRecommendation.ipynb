```python
food['Title']=food['Title'].apply(lambda x:x.lower())
```

```python
import ast
```

```python
# ast.literal_eval(food['Ingredients'])

list_of_ing=list(food['Ingredients'])

ingredients=[]
for element in list_of_ing:
  receipe=""
  for  el in ast.literal_eval(element):
    for words in el.split(" "):
      if not words.isdigit():
        receipe+=words.lower()+" "
  ingredients.append(receipe)

food['Ingredients']=ingredients
```

```python
instructions=list(food['Instructions'])
```

```python
instruct=[]
for el in instructions:
  items=""
  for words in el.split(" "):
    if not words.isdigit():
      items+=words.lower()+" "
  instruct.append(items)
food['Instructions']=instruct
```

```python
list_of_ing=list(food['Cleaned_Ingredients'])

ingredients=[]
for element in list_of_ing:
  receipe=""
  for  el in ast.literal_eval(element):
    for words in el.split(" "):
      if not words.isdigit():
        receipe+=words.lower()+" "
  ingredients.append(receipe)

food['Cleaned_Ingredients']=ingredients
```

```python
# Joining all the strings

final=[]
clean=list(food['Cleaned_Ingredients'])
ingredient=list(food['Ingredients'])
```

```python
tags=[]
for x in range(len(clean)):
  wordSet=set()
  for word in ingredient[x].split(" "):
    wordSet.add(word)
  for el in clean[x].split(" "):
    wordSet.add(el)
  
  tags.append(" ".join(list(wordSet)))

food['tags']=tags
  
```

```python
list(food['tags'][1:2])
```

```python
tags=list(food['tags'])
instruction=list(food['Instructions'])
title=list(food['Title'])
tag=[]
for x in range(len(tags)):
  tag.append(tags[x]+" "+instruction[x]+" "+title[x])

food['tags']=tag
```

```python
# for vecotorizing the words of each tags column 
from sklearn.feature_extraction.text import CountVectorizer

#  5000 words are used for feature
cv=CountVectorizer(max_features=3000, stop_words='english')
```

```python
vectors=cv.fit_transform(food['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
result=cosine_similarity(vectors)
result.shape
```

```python
similarity=result
```

```python
similarity[2].shape
```

```python
def recommend(index):

  dic={}
  
  for x in range(len(similarity[index])):
    dic[similarity[index][x]]=x
  
  descending_sorted=sorted(similarity[index],reverse=True)

  return [str(dic[j]) for j in descending_sorted[1:20]]
```

```python
food.iloc[[0]]
```

```python
indexes=recommend(4)

indexes
```

```python
recommended=[]
titles=list(food['Title'])

for x in range(len(titles)):
  recommended.append(",".join(recommend(x)))

food['recommended_to']=recommended
```

```python
titles=list(food['Title'])
indexes=[]
for x in range(len(titles)):
  indexes.append(x)
food['index']=indexes
```
