```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('movies.csv')
credits=pd.read_csv('credits.csv')
```

```python
#looking into the data
movies.head(1)
credits.head(1)
```

```python
credits.head(1)['crew']
```

```python
# merging both data frame
movies=movies.merge(credits,on="title")
movies.shape
```

```python
movies.head(1)
```

```python
# columns i want to include for machine learning
'''
genres
id (used for fetching)
keywords
title
overview
cast
crew
'''

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies
```

```python
movies.info()
```

```python
# removing the null data
movies=movies.dropna()
movies.isnull().sum()
```

```python
movies.duplicated().sum()
```

```python
# converting genres into the list
import ast

genre=list(dict(movies['genres']).values())
genres=[]
for el in genre:
  genres.append([x['name'] for x in ast.literal_eval(el)])

movies['genres']=genres
```

```python

movies.head(4)
```

```python
#  converting keywords into list
keywords=list(movies['keywords'])
names=[]
for el in keywords:
  literal_values=ast.literal_eval(el)
  names.append([el['name'] for el in literal_values])

movies['keywords']=names
```

```python
movies.head(5)
```

```python
# putting 3 cast names only
cast=list(movies['cast'])
names=[]
for el in cast:
  literal_values=ast.literal_eval(el)
  names.append(["".join(el['name'].split(" ")) for el in literal_values][:3])
names[:5]
movies['cast']=names
```

```python
movies.head(5)
```

```python
# putting director name only
print(movies['crew'].isnull().sum())
cast=list(movies['crew'])
print(len(cast))
names=[]
for el in cast:
  literal_values=ast.literal_eval(el)
  if len(literal_values)==0:
    names.append([""])
    continue
  status=False
  for details in literal_values:
    if details['job']=='Director':
      names.append(["".join(details['name'].split(" "))])
      status=True
      break
  if not status:
    print(el)
    names.append([""])
names[:5]

movies['crew']=names
```

```python
movies.head(5)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split(" "))
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python

# making tags column
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+ movies['cast']+ movies['crew']
movies.head(1)
```

```python
list(movies['tags'][:2])
```

```python
new_df=movies[['movie_id','title','tags']]
new_df
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
import nltk
```

```python
#  for changing words like loved, love, loving into same word love
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
# new_df['tags']=new_df['tags'].apply(stem)
tags_list=list(new_df['tags'])
y=[]
for el in tags_list:
  y.append(" ".join([ps.stem(word) for word in el.split(" ")]))

new_df['tags']=y
```

```python
new_df['tags'][0]
```

```python
new_df.head(1)
```

```python
# for vecotorizing the words of each tags column 
from sklearn.feature_extraction.text import CountVectorizer

#  5000 words are used for feature
cv=CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
result=cosine_similarity(vectors)
result.shape
```

```python
similarity=result
```

```python
similarity[2]
```

```python
def recommend(movie):
  movie_index=new_df[new_df['title']==movie].index[0]
  print(movie_index)
  distances=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1]) 
  return distances 
```

```python
searchFor="Superman Returns"
recommended_movies=recommend(searchFor)[1:6]

for movies in recommended_movies:
  print(new_df['title'][movies[0]])
```
